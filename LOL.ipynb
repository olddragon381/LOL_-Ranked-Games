# The impact of First Blood on victory

```python
first_blood_win = data[data['firstBlood'] == data['winner']]
first_blood_win
```

```python
fb_win_team1 = ((data['firstBlood'] == 1) & (data['winner'] == 1)).sum()
fb_total_team1 = (data['firstBlood'] == 1).sum()
fb_win_rate_team1 = (fb_win_team1 / fb_total_team1) * 100 if fb_total_team1 > 0 else 0

fb_win_team2 = ((data['firstBlood'] == 2) & (data['winner'] == 2)).sum()
fb_total_team2 = (data['firstBlood'] == 2).sum()
fb_win_rate_team2 = (fb_win_team2 / fb_total_team2) * 100 if fb_total_team2 > 0 else 0

# 📌 In kết quả
print(f"Tỷ lệ thắng khi Đội 1 có First Blood: {fb_win_rate_team1:.2f}%")
print(f"Tỷ lệ thắng khi Đội 2 có First Blood: {fb_win_rate_team2:.2f}%")
```

```python
labels = ['Đội 1 có First Blood', 'Đội 2 có First Blood']
values = [fb_win_rate_team1, fb_win_rate_team2]

plt.figure(figsize=(8, 6))
sns.barplot(x=labels, y=values, palette=['blue', 'orange'])
plt.ylabel("Tỷ lệ thắng (%)")
plt.xlabel("Đội có First Blood")
plt.title("Tỷ lệ thắng khi Đội 1 hoặc Đội 2 có First Blood", fontsize=14)
plt.ylim(0, 100)  # Giới hạn 0-100% để trực quan hơn
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, v in enumerate(values):
    plt.text(i, v + 1, f"{v:.2f}%", ha='center', fontsize=12, fontweight='bold')

plt.show()
```

# Win Prediction

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

feature=['gameDuration', 'firstTower','firstBaron',
        't1_champ1id','t1_champ2id','t1_champ3id','t1_champ4id','t1_champ5id',
        't1_towerKills','t1_inhibitorKills', 't1_baronKills', 't1_dragonKills',
        't1_riftHeraldKills', 't1_ban1', 't1_ban2', 't1_ban3', 't1_ban4','t1_ban5',
        't2_champ1id','t2_champ2id', 't2_champ3id', 't2_champ4id', 't2_champ5id',
       't2_towerKills', 't2_inhibitorKills', 't2_baronKills', 't2_dragonKills',
       't2_riftHeraldKills', 't2_ban1', 't2_ban2', 't2_ban3', 't2_ban4',
       't2_ban5']
feature_id_champ=[ 't1_champ1id','t1_champ2id','t1_champ3id','t1_champ4id','t1_champ5id', 't2_champ1id','t2_champ2id', 't2_champ3id', 't2_champ4id', 't2_champ5id','t1_ban1', 't1_ban2', 't1_ban3', 't1_ban4','t1_ban5','t2_ban1', 't2_ban2', 't2_ban3', 't2_ban4',
       't2_ban5']


X = data[feature]
y = data["winner"] - 1



scaler = StandardScaler()
cols_to_scale = [col for col in X.columns if col not in feature_id_champ]
X[cols_to_scale] = scaler.fit_transform(X[cols_to_scale])


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=24)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier

# Huấn luyện mô hình Random Forest
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Dự đoán trên tập kiểm tra
y_pred = model.predict(X_test)

# 📊 Đánh giá mô hình
print("Độ chính xác:", accuracy_score(y_test, y_pred))
print("Báo cáo phân loại:\n", classification_report(y_test, y_pred))
```

```python
from xgboost import XGBClassifier

# Khởi tạo mô hình XGBoost
xgb_model = XGBClassifier(n_estimators=100, use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.predict(X_test)

# Đánh giá độ chính xác
acc_xgb = accuracy_score(y_test, y_pred_xgb)
print(f"Độ chính xác của XGBoost: {acc_xgb:.2f}")
```
